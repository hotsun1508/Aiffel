# 6. ÏòÅÌôîÎ¶¨Î∑∞ ÌÖçÏä§Ìä∏ Í∞êÏÑ±Î∂ÑÏÑùÌïòÍ∏∞ üåºüåºüåº

## (1) ÌÖçÏä§Ìä∏Î•º Ïà´ÏûêÎ°ú ÌëúÌòÑÌïòÎäî Î∞©Î≤ï

### üåº ÏòàÏ†ú: 3Í∞úÏùò ÏßßÏùÄ Î¨∏Ïû•ÏúºÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨
>i feel hungry  
>i eat lunch  
>now i feel happy

```python
# Ï≤òÎ¶¨Ìï¥Ïïº Ìï† Î¨∏Ïû•ÏùÑ ÌååÏù¥Ïç¨ Î¶¨Ïä§Ìä∏Ïóê ÏòÆÍ≤® Îã¥ÏïòÏäµÎãàÎã§.
sentences=['i feel hungry', 'i eat lunch', 'now i feel happy']

# ÌååÏù¥Ïç¨ split() Î©îÏÜåÎìúÎ•º Ïù¥Ïö©Ìï¥ Îã®Ïñ¥ Îã®ÏúÑÎ°ú Î¨∏Ïû•ÏùÑ Ï™ºÍ∞ú Î¥ÖÎãàÎã§.
word_list = 'i feel hungry'.split()
print(word_list)
```

#### - Î™®Îì† Î¨∏Ïû•ÏùÑ Îã®Ïñ¥ Îã®ÏúÑÎ°ú Ï™ºÍ∞† ÌõÑÏóê ÌååÏù¥Ïç¨ ÎîïÏÖîÎÑàÎ¶¨(dict) ÏûêÎ£åÍµ¨Ï°∞Î°ú ÌëúÌòÑ

```python
index_to_word={}  # Îπà ÎîïÏÖîÎÑàÎ¶¨Î•º ÎßåÎì§Ïñ¥ÏÑú

# Îã®Ïñ¥Îì§ÏùÑ ÌïòÎÇòÏî© Ï±ÑÏõå Î¥ÖÎãàÎã§. Ï±ÑÏö∞Îäî ÏàúÏÑúÎäî ÏùºÎã® ÏûÑÏùòÎ°ú ÌïòÏòÄÏäµÎãàÎã§. Í∑∏Îü¨ÎÇò ÏÇ¨Ïã§ ÏàúÏÑúÎäî Ï§ëÏöîÌïòÏßÄ ÏïäÏäµÎãàÎã§. 
# <BOS>, <PAD>, <UNK>Îäî Í¥ÄÎ°ÄÏ†ÅÏúºÎ°ú ÎîïÏÖîÎÑàÎ¶¨ Îß® ÏïûÏóê ÎÑ£Ïñ¥Ï§çÎãàÎã§. 
index_to_word[0]='<PAD>'  # Ìå®Îî©Ïö© Îã®Ïñ¥
index_to_word[1]='<BOS>'  # Î¨∏Ïû•Ïùò ÏãúÏûëÏßÄÏ†ê
index_to_word[2]='<UNK>'  # ÏÇ¨Ï†ÑÏóê ÏóÜÎäî(Unknown) Îã®Ïñ¥
index_to_word[3]='i'
index_to_word[4]='feel'
index_to_word[5]='hungry'
index_to_word[6]='eat'
index_to_word[7]='lunch'
index_to_word[8]='now'
index_to_word[9]='happy'

print(index_to_word)
```

#### - ÌÖçÏä§Ìä∏Î•º Ïà´ÏûêÎ°ú Î≥ÄÌôòÌïòÍ∏∞ ÏúÑÌï¥ "ÎîïÏÖîÎÑàÎ¶¨ {ÌÖçÏä§Ìä∏:Ïù∏Îç±Ïä§}" Íµ¨Ï°∞Î°ú Î∞îÍæ∏Í∏∞

```python
word_to_index={word:index for index, word in index_to_word.items()}
print(word_to_index)
```

```python
# Ïù¥ ÎîïÏÖîÎÑàÎ¶¨Îäî Îã®Ïñ¥Î•º Ï£ºÎ©¥ Í∑∏ Îã®Ïñ¥Ïùò Ïù∏Îç±Ïä§Î•º Î∞òÌôòÌïòÎäî Î∞©ÏãùÏúºÎ°ú ÏÇ¨Ïö© Í∞ÄÎä•
print(word_to_index['feel'])  # Îã®Ïñ¥ 'feel'ÏùÄ Ïà´Ïûê Ïù∏Îç±Ïä§ 4Î°ú Î∞îÎÄùÎãàÎã§.
```

#### - ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ïà´ÏûêÎ°ú Î∞îÍøî ÌëúÌòÑÌï¥Î≥¥Í∏∞

```python
# Î¨∏Ïû• 1Í∞úÎ•º ÌôúÏö©Ìï† ÎîïÏÖîÎÑàÎ¶¨ÏôÄ Ìï®Íªò Ï£ºÎ©¥, Îã®Ïñ¥ Ïù∏Îç±Ïä§ Î¶¨Ïä§Ìä∏Î°ú Î≥ÄÌôòÌï¥ Ï£ºÎäî Ìï®ÏàòÎ•º ÎßåÎì§Ïñ¥ Î¥ÖÏãúÎã§.
# Îã®, Î™®Îì† Î¨∏Ïû•ÏùÄ <BOS>Î°ú ÏãúÏûëÌïòÎäî Í≤ÉÏúºÎ°ú Ìï©ÎãàÎã§. 
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

print(get_encoded_sentence('i eat lunch', word_to_index))
```

#### - get_encoded_sentence Ìï®ÏàòÎ•º ÌÜµÌï¥ ÏïÑÎûòÏôÄ Í∞ôÏù¥ Î¨∏Ïû• Ïó¨Îü¨Í∞ú ÌïúÎ≤àÏóê Îß§ÌïëÌïòÍ∏∞

> `<BOS>` -> 1  
> i -> 3  
> eat -> 6  
> lunch -> 7  

```python
# Ïó¨Îü¨ Í∞úÏùò Î¨∏Ïû• Î¶¨Ïä§Ìä∏Î•º ÌïúÍ∫ºÎ≤àÏóê Ïà´Ïûê ÌÖêÏÑúÎ°ú encodeÌï¥ Ï£ºÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# sentences=['i feel hungry', 'i eat lunch', 'now i feel happy'] Í∞Ä ÏïÑÎûòÏôÄ Í∞ôÏù¥ Î≥ÄÌôòÎê©ÎãàÎã§. 
encoded_sentences = get_encoded_sentences(sentences, word_to_index)
print(encoded_sentences)
```

#### - Î∞òÎåÄÎ°ú, encodeÎêú Î≤°ÌÑ∞Î•º decodeÌïòÏó¨ Îã§Ïãú ÏõêÎûò ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î°ú Î≥µÍµ¨ÌïòÍ∏∞

```python
# Ïà´Ïûê Î≤°ÌÑ∞Î°ú encodeÎêú Î¨∏Ïû•ÏùÑ ÏõêÎûòÎåÄÎ°ú decodeÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]Î•º ÌÜµÌï¥ <BOS>Î•º Ï†úÏô∏

print(get_decoded_sentence([1, 3, 4, 5], index_to_word))
```

```python
# Ïó¨Îü¨ Í∞úÏùò Ïà´Ïûê Î≤°ÌÑ∞Î°ú encodeÎêú Î¨∏Ïû•ÏùÑ ÌïúÍ∫ºÎ≤àÏóê ÏõêÎûòÎåÄÎ°ú decodeÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]

# encoded_sentences=[[1, 3, 4, 5], [1, 3, 6, 7], [1, 8, 3, 4, 9]] Í∞Ä ÏïÑÎûòÏôÄ Í∞ôÏù¥ Î≥ÄÌôòÎê©ÎãàÎã§.
print(get_decoded_sentences(encoded_sentences, index_to_word))
```

## (2) Embedding Î†àÏù¥Ïñ¥Ïùò Îì±Ïû•

## üåºÏö∞Î¶¨Í∞Ä ÌïòÎ†§Îäî Í≤É:  Îã®Ïñ¥ÏôÄ Í∑∏ Îã®Ïñ¥Ïùò ÏùòÎØ∏Î•º ÎÇòÌÉÄÎÇ¥Îäî Î≤°ÌÑ∞Î•º ÏßùÏßìÎäî Í≤É  
> **Îã®Ïñ¥Ïùò ÏùòÎØ∏Î•º ÎÇòÌÉÄÎÇ¥Îäî Î≤°ÌÑ∞ = ÌõàÎ†® Í∞ÄÎä•Ìïú ÌååÎùºÎØ∏ÌÑ∞**Î°ú ÎÜìÍ≥† Ïù¥Î•º Îî•Îü¨ÎãùÏùÑ ÌÜµÌï¥ ÌïôÏäµÌï¥ÏÑú ÏµúÏ†ÅÌôîÌïòÍ≤å Îê©ÎãàÎã§.   
> `Tensorflow, Pytorch` Îì±Ïùò Îî•Îü¨Îãù ÌîÑÎ†àÏûÑÏõåÌÅ¨Îì§ÏùÄ Ïù¥Îü¨Ìïú **ÏùòÎØ∏ Î≤°ÌÑ∞ ÌååÎùºÎØ∏ÌÑ∞**Î•º Íµ¨ÌòÑÌïú `Embedding Î†àÏù¥Ïñ¥`Î•º Ï†úÍ≥µÌï©ÎãàÎã§.

```python
# ÏïÑÎûò ÏΩîÎìúÎäî Í∑∏ÎåÄÎ°ú Ïã§ÌñâÌïòÏãúÎ©¥ ÏóêÎü¨Í∞Ä Î∞úÏÉùÌï† Í≤ÉÏûÖÎãàÎã§. 

import numpy as np
import tensorflow as tf
import os

vocab_size = len(word_to_index)  # ÏúÑ ÏòàÏãúÏóêÏÑú ÎîïÏÖîÎÑàÎ¶¨Ïóê Ìè¨Ìï®Îêú Îã®Ïñ¥ Í∞úÏàòÎäî 10
word_vector_dim = 4    # ÏúÑ Í∑∏Î¶ºÍ≥º Í∞ôÏù¥ 4Ï∞®ÏõêÏùò ÏõåÎìú Î≤°ÌÑ∞Î•º Í∞ÄÏ†ïÌï©ÎãàÎã§. 

embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=word_vector_dim, mask_zero=True)

# Ïà´ÏûêÎ°ú Î≥ÄÌôòÎêú ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ [[1, 3, 4, 5], [1, 3, 6, 7], [1, 8, 3, 4, 9]] Ïóê Embedding Î†àÏù¥Ïñ¥Î•º Ï†ÅÏö©Ìï©ÎãàÎã§. 
raw_inputs = np.array(get_encoded_sentences(sentences, word_to_index), dtype='object')
output = embedding(raw_inputs)
print(output)
```

## ü¶ÄKey Pointü¶Ä
>Embedding Î†àÏù¥Ïñ¥Ïùò Ïù∏ÌíãÏù¥ ÎêòÎäî Î¨∏Ïû• Î≤°ÌÑ∞Îäî Í∑∏ Í∏∏Ïù¥Í∞Ä ÏùºÏ†ïÌï¥Ïïº ÌïúÎã§.   
>raw_inputsÏùò 3Í∞ú Î≤°ÌÑ∞Ïùò Í∏∏Ïù¥Îäî Í∞ÅÍ∞Å 4, 4, 5.

#### - TensorflowÏùò`tf.keras.preprocessing.sequence.pad_sequences` (Í∏∞Îä•)Ìï®ÏàòÎ•º ÌÜµÌï¥ Î¨∏Ïû• Î≤°ÌÑ∞ Îí§Ïóê Ìå®Îî©(`<PAD>`)ÏùÑ Ï∂îÍ∞ÄÌïòÏó¨ Í∏∏Ïù¥ ÏùºÏ†ïÌïòÍ≤å ÎßûÏ∂∞Ï£ºÍ∏∞

```python
raw_inputs = tf.keras.preprocessing.sequence.pad_sequences(raw_inputs,
                                                       value=word_to_index['<PAD>'], # <PAD>Îäî 0Ïóê Îß§ÌïëÎêòÏñ¥ ÏûàÏùå
                                                       padding='post',
                                                       maxlen=5)
print(raw_inputs)
```

#### ÏßßÏùÄ Î¨∏Ïû• Îí§Ï™ΩÏù¥ 0ÏúºÎ°ú Ï±ÑÏõåÏßÄÎäî Í≤ÉÏùÑ ÌôïÏù∏ !! 
#### - output = embedding(raw_inputs) Îã§Ïãú ÏãúÎèÑ

```python
vocab_size = len(word_to_index)  # ÏúÑ ÏòàÏãúÏóêÏÑú ÎîïÏÖîÎÑàÎ¶¨Ïóê Ìè¨Ìï®Îêú Îã®Ïñ¥ Í∞úÏàòÎäî 10
word_vector_dim = 4    # Í∑∏Î¶ºÍ≥º Í∞ôÏù¥ 4Ï∞®ÏõêÏùò ÏõåÎìú Î≤°ÌÑ∞Î•º Í∞ÄÏ†ïÌï©ÎãàÎã§.

embedding = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=word_vector_dim, mask_zero=True)

# tf.keras.preprocessing.sequence.pad_sequencesÎ•º ÌÜµÌï¥ word vectorÎ•º Î™®Îëê ÏùºÏ†ï Í∏∏Ïù¥Î°ú ÎßûÏ∂∞Ï£ºÏñ¥Ïïº 
# embedding Î†àÏù¥Ïñ¥Ïùò inputÏù¥ Îê† Ïàò ÏûàÏùåÏóê Ï£ºÏùòÌï¥ Ï£ºÏÑ∏Ïöî. 
raw_inputs = np.array(get_encoded_sentences(sentences, word_to_index), dtype=object)
raw_inputs = tf.keras.preprocessing.sequence.pad_sequences(raw_inputs,
                                                       value=word_to_index['<PAD>'],
                                                       padding='post',
                                                       maxlen=5)
output = embedding(raw_inputs)
print(output)
```

## (3) ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞Î•º Îã§Î£®Îäî RNN
> üåº ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Îã§Î£®Îäî Îç∞ Ï£ºÎ°ú ÏÇ¨Ïö©ÎêòÎäî Îî•Îü¨Îãù Î™®Îç∏ÏùÄ Î∞îÎ°ú Recurrent Neural Network(RNN)ÏûÖÎãàÎã§.   
    RNNÏùÄ ÏãúÌÄÄÏä§(Sequence) ÌòïÌÉúÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨ÌïòÍ∏∞Ïóê ÏµúÏ†ÅÏù∏ Î™®Îç∏Î°ú ÏïåÎ†§Ï†∏ ÏûàÏäµÎãàÎã§.

- RNNÏùÄ ÏãúÍ∞ÑÏùò ÌùêÎ¶ÑÏóê Îî∞Îùº ÏÉàÎ°≠Í≤å Îì§Ïñ¥Ïò§Îäî ÏûÖÎ†•Ïóê Îî∞Îùº Î≥ÄÌïòÎäî ÌòÑÏû¨ ÏÉÅÌÉúÎ•º Î¨òÏÇ¨ÌïòÎäî state machineÏúºÎ°ú ÏÑ§Í≥ÑÎêòÏóàÎã§.
= Stateful Ìï® !! 





**Ï∞∏Í≥† ÏòÅÏÉÅ: [ÍπÄÏÑ±Ìõà ÍµêÏàòÎãò Î™®ÎëêÏùò Îî•Îü¨Îãù 12Í∞ï RNN](https://youtu.be/-SHPG_KMUkQ)üåº**
 **[Í∞êÏ†ï Î∂ÑÏÑùÏóêÏÑú RNN ÌôúÏö©]**


                                                

#### - RNN Î™®Îç∏ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Ïù¥Ï†Ñ Ïä§ÌÖùÏùò ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨ÌïòÎäî ÏòàÏ†ú ÏΩîÎìúÎ•º Íµ¨ÌòÑÌï¥Î≥¥Í∏∞

```python
vocab_size = 10  # Ïñ¥Ìúò ÏÇ¨Ï†ÑÏùò ÌÅ¨Í∏∞ÏûÖÎãàÎã§(10Í∞úÏùò Îã®Ïñ¥)
word_vector_dim = 4  # Îã®Ïñ¥ ÌïòÎÇòÎ•º ÌëúÌòÑÌïòÎäî ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Ïùò Ï∞®ÏõêÏàòÏûÖÎãàÎã§. 

model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model.add(tf.keras.layers.LSTM(8))   # Í∞ÄÏû• ÎÑêÎ¶¨ Ïì∞Ïù¥Îäî RNNÏù∏ LSTM Î†àÏù¥Ïñ¥Î•º ÏÇ¨Ïö©ÌïòÏòÄÏäµÎãàÎã§. Ïù¥Îïå LSTM state Î≤°ÌÑ∞Ïùò Ï∞®ÏõêÏàòÎäî 8Î°ú ÌïòÏòÄÏäµÎãàÎã§. (Î≥ÄÍ≤Ω Í∞ÄÎä•)
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # ÏµúÏ¢Ö Ï∂úÎ†•ÏùÄ Í∏çÏ†ï/Î∂ÄÏ†ïÏùÑ ÎÇòÌÉÄÎÇ¥Îäî 1dim ÏûÖÎãàÎã§.

model.summary()
```

## (4) Íº≠ RNNÏù¥Ïñ¥Ïïº Ìï†Íπå? 

üåπYES! 1-D Convolution Neural Network(1-D CNN) ÏÇ¨Ïö©
-  Î¨∏Ïû• Ï†ÑÏ≤¥Î•º ÌïúÍ∫ºÎ≤àÏóê Ìïú Î∞©Ìñ•ÏúºÎ°ú Í∏∏Ïù¥ 7ÏßúÎ¶¨ ÌïÑÌÑ∞Î°ú Ïä§Ï∫êÎãù ÌïòÎ©¥ÏÑú 7Îã®Ïñ¥ Ïù¥ÎÇ¥ÏóêÏÑú Î∞úÍ≤¨ÎêòÎäî ÌäπÏßïÏùÑ Ï∂îÏ∂úÌïòÏó¨ Í∑∏Í≤ÉÏúºÎ°ú Î¨∏Ïû•ÏùÑ Î∂ÑÎ•òÌïòÎäî Î∞©ÏãùÏúºÎ°ú ÏÇ¨Ïö©
- CNN Í≥ÑÏó¥ÏùÄ RNN Í≥ÑÏó¥Î≥¥Îã§ Î≥ëÎ†¨Ï≤òÎ¶¨Í∞Ä Ìö®Ïú®Ï†ÅÏù¥Í∏∞ ÎïåÎ¨∏Ïóê ÌïôÏäµ ÏÜçÎèÑÎèÑ Ìõ®Ïî¨ Îπ†Î•¥Í≤å ÏßÑÌñâÎêúÎã§

ü•ÄNO! 2-D CNN: Ïù¥ÎØ∏ÏßÄ Î∂ÑÎ•òÍ∏∞ Î™®Îç∏ Íµ¨ÌòÑ Ïãú ÏÇ¨Ïö©. Ïù¥ÎØ∏ÏßÄÎäî ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÎãàÍ∏∞ ÎïåÎ¨∏Ïóê ÏïÑÎãàÎùº Ïù¥ÎØ∏ÏßÄ Ï†ÑÏ≤¥Í∞Ä ÌïúÍ∫ºÎ≤àÏóê ÏûÖÎ†•ÏúºÎ°ú ÏÇ¨Ïö©Îê®. 

```python
vocab_size = 10  # Ïñ¥Ìúò ÏÇ¨Ï†ÑÏùò ÌÅ¨Í∏∞ÏûÖÎãàÎã§(10Í∞úÏùò Îã®Ïñ¥)
word_vector_dim = 4   # Îã®Ïñ¥ ÌïòÎÇòÎ•º ÌëúÌòÑÌïòÎäî ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Ïùò Ï∞®Ïõê ÏàòÏûÖÎãàÎã§. 

model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model.add(tf.keras.layers.MaxPooling1D(5))
model.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model.add(tf.keras.layers.GlobalMaxPooling1D())
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # ÏµúÏ¢Ö Ï∂úÎ†•ÏùÄ Í∏çÏ†ï/Î∂ÄÏ†ïÏùÑ ÎÇòÌÉÄÎÇ¥Îäî 1dim ÏûÖÎãàÎã§.

model.summary()
```

#### - GlobalMaxPooling1D() Î†àÏù¥Ïñ¥ ÌïòÎÇòÎßå ÏÇ¨Ïö©Ìï¥Î≥¥Í∏∞
> Ï†ÑÏ≤¥ Î¨∏Ïû• Ï§ëÏóêÏÑú Îã® ÌïòÎÇòÏùò Í∞ÄÏû• Ï§ëÏöîÌïú Îã®Ïñ¥Îßå ÌîºÏ≤òÎ°ú Ï∂îÏ∂úÌïòÏó¨ Í∑∏Í≤ÉÏúºÎ°ú Î¨∏Ïû•Ïùò Í∏çÏ†ï/Î∂ÄÏ†ïÏùÑ ÌèâÍ∞ÄÌïòÎäî Î∞©Ïãù

```python
vocab_size = 10  # Ïñ¥Ìúò ÏÇ¨Ï†ÑÏùò ÌÅ¨Í∏∞ÏûÖÎãàÎã§(10Í∞úÏùò Îã®Ïñ¥)
word_vector_dim = 4   # Îã®Ïñ¥ ÌïòÎÇòÎ•º ÌëúÌòÑÌïòÎäî ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Ïùò Ï∞®Ïõê ÏàòÏûÖÎãàÎã§. 

model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model.add(tf.keras.layers.GlobalMaxPooling1D())
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # ÏµúÏ¢Ö Ï∂úÎ†•ÏùÄ Í∏çÏ†ï/Î∂ÄÏ†ïÏùÑ ÎÇòÌÉÄÎÇ¥Îäî 1dim ÏûÖÎãàÎã§.

model.summary()
```

### üåºÎã§ÏñëÌïú ÏãúÎèÑ Í∞ÄÎä•
- 1-D CNNÍ≥º RNN Î†àÏù¥Ïñ¥Î•º ÏÑûÏñ¥ Ïì∞Í∏∞
- FFN(FeedForward Network) Î†àÏù¥Ïñ¥ÎßåÏúºÎ°ú Íµ¨ÏÑ±ÌïòÍ∏∞
- Transformer Î†àÏù¥Ïñ¥Î•º Ïç®Î≥¥Í∏∞

> imdb.load_data() Ìò∏Ï∂ú Ïãú Îã®Ïñ¥ÏÇ¨Ï†ÑÏóê Îì±Ïû¨Ìï† Îã®Ïñ¥Ïùò Í∞úÏàò(num_words)Î•º 10000ÏúºÎ°ú ÏßÄÏ†ïÌïòÎ©¥, Í∑∏ Í∞úÏàòÎßåÌÅºÏùò word_to_index ÎîïÏÖîÎÑàÎ¶¨ÍπåÏßÄ ÏÉùÏÑ±Îêú ÌòïÌÉúÎ°ú Îç∞Ïù¥ÌÑ∞ÏÖãÏù¥ ÏÉùÏÑ±

```python
imdb = tf.keras.datasets.imdb

# IMDb Îç∞Ïù¥ÌÑ∞ÏÖã Îã§Ïö¥Î°úÎìú 
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000) # Îã®Ïñ¥ÏÇ¨Ï†ÑÏóê Îì±Ïû¨Ìï† Îã®Ïñ¥Ïùò Í∞úÏàò
print("ÌõàÎ†® ÏÉòÌîå Í∞úÏàò: {}, ÌÖåÏä§Ìä∏ Í∞úÏàò: {}".format(len(x_train), len(x_test)))
```

#### - Îã§Ïö¥Î°úÎìúÌïú Îç∞Ïù¥ÌÑ∞ Ïã§Ï†ú ÏòàÏãú ÌôïÏù∏Ìï¥Î≥¥Í∏∞

```python
print(x_train[0])  # 1Î≤àÏß∏ Î¶¨Î∑∞Îç∞Ïù¥ÌÑ∞
print('ÎùºÎ≤®: ', y_train[0])  # 1Î≤àÏß∏ Î¶¨Î∑∞Îç∞Ïù¥ÌÑ∞Ïùò ÎùºÎ≤®
print('1Î≤àÏß∏ Î¶¨Î∑∞ Î¨∏Ïû• Í∏∏Ïù¥: ', len(x_train[0]))
print('2Î≤àÏß∏ Î¶¨Î∑∞ Î¨∏Ïû• Í∏∏Ïù¥: ', len(x_train[1]))
```

>ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÎãàÎùº Ïù¥ÎØ∏ Ïà´ÏûêÎ°ú encodeÎêú Îç∞Ïù¥ÌÑ∞Î°ú Ï∂úÎ†•.   
>IMDb Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÎäî encodeÏóê ÏÇ¨Ïö©Ìïú ÎîïÏÖîÎÑàÎ¶¨ÍπåÏßÄ Ìï®Íªò Ï†úÍ≥µÌï® 

```python
word_to_index = imdb.get_word_index()
index_to_word = {index:word for word, index in word_to_index.items()}
print(index_to_word[1])     # 'the' Í∞Ä Ï∂úÎ†•Îê©ÎãàÎã§. 
print(word_to_index['the'])  # 1 Ïù¥ Ï∂úÎ†•Îê©ÎãàÎã§.
```

#### - MDb Îç∞Ïù¥ÌÑ∞ÏÖãÏùò ÌÖçÏä§Ìä∏ Ïù∏ÏΩîÎî©ÏùÑ ÏúÑÌïú word_to_index, index_to_wordÎäî ÏïÑÎûòÏôÄ Í∞ôÏù¥ Î≥¥Ï†ï
> word_to_indexÎäî IMDb ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖãÏùò Îã®Ïñ¥ Ï∂úÌòÑ ÎπàÎèÑ Í∏∞Ï§ÄÏúºÎ°ú ÎÇ¥Î¶ºÏ∞®ÏàúÏúºÎ°ú Ï†ïÎ†¨Îê® 

```python
#Ïã§Ï†ú Ïù∏ÏΩîÎî© Ïù∏Îç±Ïä§Îäî Ï†úÍ≥µÎêú word_to_indexÏóêÏÑú index Í∏∞Ï§ÄÏúºÎ°ú 3Ïî© Îí§Î°ú Î∞ÄÎ†§ ÏûàÏäµÎãàÎã§.  
word_to_index = {k:(v+3) for k,v in word_to_index.items()}

# Ï≤òÏùå Î™á Í∞ú Ïù∏Îç±Ïä§Îäî ÏÇ¨Ï†ÑÏóê Ï†ïÏùòÎêòÏñ¥ ÏûàÏäµÎãàÎã§
word_to_index["<PAD>"] = 0
word_to_index["<BOS>"] = 1
word_to_index["<UNK>"] = 2  # unknown
word_to_index["<UNUSED>"] = 3

index_to_word = {index:word for word, index in word_to_index.items()}

print(index_to_word[1])     # '<BOS>' Í∞Ä Ï∂úÎ†•Îê©ÎãàÎã§. 
print(word_to_index['the'])  # 4 Ïù¥ Ï∂úÎ†•Îê©ÎãàÎã§. 
print(index_to_word[4])     # 'the' Í∞Ä Ï∂úÎ†•Îê©ÎãàÎã§.
```

#### -  encodeÎêú ÌÖçÏä§Ìä∏Í∞Ä Ï†ïÏÉÅÏ†ÅÏúºÎ°ú decode ÎêòÎäîÏßÄ ÌôïÏù∏Ìï¥Î≥¥Í∏∞

```python
print(get_decoded_sentence(x_train[0], index_to_word))
print('ÎùºÎ≤®: ', y_train[0])  # 1Î≤àÏß∏ Î¶¨Î∑∞Îç∞Ïù¥ÌÑ∞Ïùò ÎùºÎ≤®
```

#### - pad_sequencesÎ•º ÌÜµÌï¥ Îç∞Ïù¥ÌÑ∞ÏÖã ÏÉÅÏùò Î¨∏Ïû•Ïùò Í∏∏Ïù¥ ÌÜµÏùºÌïòÍ∏∞
> Î¨∏Ïû• ÏµúÎåÄ Í∏∏Ïù¥ maxlenÏùò Ï†ÅÏ†àÌïú Í∞í ÏÑ§Ï†ï -> Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÏÖãÏùò Î∂ÑÌè¨ ÌôïÏù∏ 

```python
total_data_text = list(x_train) + list(x_test)
# ÌÖçÏä§Ìä∏Îç∞Ïù¥ÌÑ∞ Î¨∏Ïû•Í∏∏Ïù¥Ïùò Î¶¨Ïä§Ìä∏Î•º ÏÉùÏÑ±Ìïú ÌõÑ
num_tokens = [len(tokens) for tokens in total_data_text]
num_tokens = np.array(num_tokens)
# Î¨∏Ïû•Í∏∏Ïù¥Ïùò ÌèâÍ∑†Í∞í, ÏµúÎåÄÍ∞í, ÌëúÏ§ÄÌé∏Ï∞®Î•º Í≥ÑÏÇ∞Ìï¥ Î≥∏Îã§. 
print('Î¨∏Ïû•Í∏∏Ïù¥ ÌèâÍ∑† : ', np.mean(num_tokens))
print('Î¨∏Ïû•Í∏∏Ïù¥ ÏµúÎåÄ : ', np.max(num_tokens))
print('Î¨∏Ïû•Í∏∏Ïù¥ ÌëúÏ§ÄÌé∏Ï∞® : ', np.std(num_tokens))

# ÏòàÎ•ºÎì§Ïñ¥, ÏµúÎåÄ Í∏∏Ïù¥Î•º (ÌèâÍ∑† + 2*ÌëúÏ§ÄÌé∏Ï∞®)Î°ú ÌïúÎã§Î©¥,  
max_tokens = np.mean(num_tokens) + 2 * np.std(num_tokens)
maxlen = int(max_tokens)
print('pad_sequences maxlen : ', maxlen)  # Ïù¥ Í≤ΩÏö∞ÏóêÎäî maxlen=580
print('Ï†ÑÏ≤¥ Î¨∏Ïû•Ïùò {}%Í∞Ä maxlen ÏÑ§Ï†ïÍ∞í Ïù¥ÎÇ¥Ïóê Ìè¨Ìï®Îê©ÎãàÎã§. '.format(np.sum(num_tokens < max_tokens) / len(num_tokens)))
```

#### - Padding Î∞©Ïãù ÏÑ§Ï†ïÌïòÍ∏∞: Îëê Í∞ÄÏßÄ Î∞©Ïãù Î™®Îëê Îã§ Ï†ÅÏö©ÌïòÍ∏∞
- Î¨∏Ïû• Îí§Ï™Ω('post')Í≥º ÏïûÏ™Ω('pre') Ï§ë ÌÉùÌïòÍ∏∞  
- RNNÏùÑ Ïù¥Ïö©Ìïú Îî•Îü¨Îãù Ï†ÅÏö© Ïãú ÏÑ±Îä• Ï∞®Ïù¥ Î∞úÏÉù 
> üåº RNNÏùÄ ÏûÖÎ†•Îç∞Ïù¥ÌÑ∞Í∞Ä ÏàúÏ∞®Ï†ÅÏúºÎ°ú Ï≤òÎ¶¨ÎêòÏñ¥, ÎßàÏßÄÎßâ ÏûÖÎ†•Ïù¥ ÏµúÏ¢Ö state Í∞íÏóê ÏòÅÌñ•ÏùÑ Í∞ÄÏû• ÎßéÏù¥ ÎØ∏Ïπ®
> - ÎßàÏßÄÎßâ ÏûÖÎ†•Ïù¥ Î¨¥ÏùòÎØ∏Ìïú paddingÏúºÎ°ú Ï±ÑÏõåÏßÄÎäî Í≤ÉÏùÄ ÎπÑÌö®Ïú®Ï†Å
> - 'pre'Í∞Ä Ìõ®Ïî¨ Ïú†Î¶¨ÌïòÎ©∞, 10% Ïù¥ÏÉÅÏùò ÌÖåÏä§Ìä∏ ÏÑ±Îä• Ï∞®Ïù¥Î•º Î≥¥ÏûÑ.

```python
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train,
                                                        value=word_to_index["<PAD>"],
                                                        padding='post', # ÌòπÏùÄ 'pre'
                                                        maxlen=maxlen)

x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test,
                                                       value=word_to_index["<PAD>"],
                                                       padding='post', # ÌòπÏùÄ 'pre'
                                                       maxlen=maxlen)

print(x_train.shape)
```

## (2) Îî•Îü¨Îãù Î™®Îç∏ ÏÑ§Í≥ÑÏôÄ ÌõàÎ†®
#### - RNN Î™®Îç∏ ÏÑ§Í≥ÑÌï¥Î≥¥Í∏∞ (Îã§ÏñëÌïú Î™®Îç∏Î°ú Íµ¨ÌòÑ Í∞ÄÎä•)

```python
vocab_size = 10000    # Ïñ¥Ìúò ÏÇ¨Ï†ÑÏùò ÌÅ¨Í∏∞ÏûÖÎãàÎã§(10,000Í∞úÏùò Îã®Ïñ¥)
word_vector_dim = 16  # ÏõåÎìú Î≤°ÌÑ∞Ïùò Ï∞®Ïõê Ïàò (Î≥ÄÍ≤Ω Í∞ÄÎä•Ìïú ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞)

# model ÏÑ§Í≥Ñ - Îî•Îü¨Îãù Î™®Îç∏ ÏΩîÎìúÎ•º ÏßÅÏ†ë ÏûëÏÑ±Ìï¥ Ï£ºÏÑ∏Ïöî.
model = tf.keras.Sequential()
# [[YOUR CODE]]
model.add(tf.keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model.add(tf.keras.layers.LSTM(8))   # Í∞ÄÏû• ÎÑêÎ¶¨ Ïì∞Ïù¥Îäî RNNÏù∏ LSTM Î†àÏù¥Ïñ¥ ÏÇ¨Ïö©. LSTM state Î≤°ÌÑ∞Ïùò Ï∞®ÏõêÏàòÎäî 8 (Î≥ÄÍ≤Ω Í∞ÄÎä•)
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))  # ÏµúÏ¢Ö Ï∂úÎ†•ÏùÄ Í∏çÏ†ï/Î∂ÄÏ†ïÏùÑ ÎÇòÌÉÄÎÇ¥Îäî 1dim ÏûÖÎãàÎã§.

model.summary()
```

#### - ÌõàÎ†®Ïö© Îç∞Ïù¥ÌÑ∞ÏÖã 25000Í±¥ Ï§ë 10000Í±¥ÏùÑ Î∂ÑÎ¶¨ÌïòÏó¨ Í≤ÄÏ¶ùÏÖã(validation set)ÏúºÎ°ú ÏÇ¨Ïö©ÌïòÍ∏∞
> - model ÌõàÎ†® Ï†Ñ Îã®Í≥Ñ  
> - Ï†ÅÏ†àÌïú validation Îç∞Ïù¥ÌÑ∞Îäî Î™á Í∞úÏùºÏßÄ Í≥†ÎØº

```python
# validation set 10000Í±¥ Î∂ÑÎ¶¨
x_val = x_train[:10000]   
y_val = y_train[:10000]

# validation setÏùÑ Ï†úÏô∏Ìïú ÎÇòÎ®∏ÏßÄ 15000Í±¥
partial_x_train = x_train[10000:]  
partial_y_train = y_train[10000:]

print(partial_x_train.shape)
print(partial_y_train.shape)
```

#### - model ÌïôÏäµ ÏãúÏûëÌïòÍ∏∞
> `model.fit()` Í≥ºÏ†ï Ï§ëÏùò train/validation loss, accuracy Îì±Ïù¥ Îß§ epochÎßàÎã§ history Î≥ÄÏàòÏóê Ï†ÄÏû•ÎêòÏñ¥ ÏûàÏùå

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
              
epochs=20  # Î™á epochÎ•º ÌõàÎ†®ÌïòÎ©¥ Ï¢ãÏùÑÏßÄ Í≤∞Í≥ºÎ•º Î≥¥Î©¥ÏÑú Î∞îÍæ∏Ïñ¥ Î¥ÖÏãúÎã§. 

history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)
```

#### - ÌïôÏäµÏù¥ ÎÅùÎÇú Î™®Îç∏ÏùÑ ÌÖåÏä§Ìä∏ÏÖãÏúºÎ°ú ÌèâÍ∞ÄÌïòÍ∏∞

```python
results = model.evaluate(x_test,  y_test, verbose=2)

print(results)
```

#### - Îç∞Ïù¥ÌÑ∞Î•º Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î†§Î≥¥Í∏∞
> üåº Îã§ÏñëÌïú ÏïÑÏù¥ÎîîÏñ¥ ÏñªÍ∏∞
> - Îî•Îü¨Îãù ÌïôÏäµÏù¥ Ïûò ÏßÑÌñâÎêòÏóàÎäîÏßÄ
> - Ïò§Î≤ÑÌîºÌåÖ ÌòπÏùÄ Ïñ∏ÎçîÌîºÌåÖÌïòÏßÄ ÏïäÏïòÎäîÏßÄ 
> - ÏÑ±Îä• Í∞úÏÑ† Î∞©Î≤ï

```python
history_dict = history.history
print(history_dict.keys()) # epochÏóê Îî∞Î•∏ Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Î≥º Ïàò ÏûàÎäî Ìï≠Î™©Îì§
```

1. Training and validation loss Í∑∏ÎûòÌîÑ
>- Î™á epochÍπåÏßÄÏùò Ìä∏Î†àÏù¥ÎãùÏù¥ Ï†ÅÏ†àÌïúÏßÄ ÏµúÏ†ÅÏ†ê Ï∂îÏ†ïÌï¥Î≥¥Í∏∞. 
>- validation lossÏùò Í∑∏ÎûòÌîÑÍ∞Ä train lossÏôÄÏùò Ïù¥Í≤©Ïù¥ Î∞úÏÉùÌïòÍ≤å ÎêòÎ©¥ Îçî Ïù¥ÏÉÅÏùò Ìä∏Î†àÏù¥ÎãùÏùÄ Î¨¥ÏùòÎØ∏Ìï®

```python
import matplotlib.pyplot as plt

acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)

# "bo"Îäî "ÌååÎûÄÏÉâ Ï†ê"ÏûÖÎãàÎã§
plt.plot(epochs, loss, 'bo', label='Training loss')
# bÎäî "ÌååÎûÄ Ïã§ÏÑ†"ÏûÖÎãàÎã§
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

## üåº ÏõåÎìú ÏûÑÎ≤†Îî©(word embedding) Í∏∞Î≤ï
ÎùºÎ≤®ÎßÅ ÎπÑÏö©Ïù¥ ÎßéÏù¥ ÎìúÎäî Î®∏Ïã†Îü¨Îãù Í∏∞Î∞ò Í∞êÏÑ±Î∂ÑÏÑùÏùò ÎπÑÏö©ÏùÑ Ï†àÍ∞êÌïòÎ©¥ÏÑú Ï†ïÌôïÎèÑÎ•º ÌÅ¨Í≤å Ìñ•ÏÉÅÏãúÌÇ¨ Ïàò ÏûàÎäî ÏûêÏó∞Ïñ¥Ï≤òÎ¶¨ Í∏∞Î≤ïÏúºÎ°ú Îã®Ïñ¥Ïùò ÌäπÏÑ±ÏùÑ Ï†ÄÏ∞®Ïõê Î≤°ÌÑ∞Í∞íÏúºÎ°ú ÌëúÌòÑÌïúÎã§. 

> **ü¶Ä Embedding Ïù¥ÎûÄ?**  
> ÏûÑÎ≤†Îî©(embedding)ÏùÄ ÏûêÏó∞Ïñ¥Î•º Ïà´ÏûêÏùò ÎÇòÏó¥Ïù∏ Î≤°ÌÑ∞Î°ú Î∞îÍæº Í≤∞Í≥º ÌòπÏùÄ Í∑∏ ÏùºÎ†®Ïùò Í≥ºÏ†ï Ï†ÑÏ≤¥Î•º Í∞ÄÎ¶¨ÌÇ§Îäî Ïö©Ïñ¥ÏûÖÎãàÎã§. Îã®Ïñ¥ÎÇò Î¨∏Ïû• Í∞ÅÍ∞ÅÏùÑ Î≤°ÌÑ∞Î°ú Î≥ÄÌôòÌï¥ Î≤°ÌÑ∞ Í≥µÍ∞ÑÏóê ‚ÄòÎÅºÏõå ÎÑ£ÎäîÎã§(embed)‚ÄôÎäî Ï∑®ÏßÄÏóêÏÑú ÏûÑÎ≤†Îî©Ïù¥ÎùºÎäî Ïù¥Î¶ÑÏù¥ Î∂ôÏóàÏäµÎãàÎã§. Ïª¥Ìì®ÌÑ∞Í∞Ä ÏûêÏó∞Ïñ¥Î•º Ï≤òÎ¶¨Ìï† Ïàò ÏûàÍ≤å ÌïòÎ†§Î©¥ ÏûêÏó∞Ïñ¥Î•º Í≥ÑÏÇ∞ Í∞ÄÎä•Ìïú ÌòïÏãùÏù∏ ÏûÑÎ≤†Îî©ÏúºÎ°ú Î∞îÍøîÏ§òÏïº Ìï©ÎãàÎã§.   

> **ü¶ÄÏ†ÑÏù¥ ÌïôÏäµÏù¥ÎûÄ?**   
> ÌäπÏ†ï Î¨∏Ï†úÎ•º ÌíÄÍ∏∞ ÏúÑÌï¥ ÌïôÏäµÌïú Î™®Îç∏ÏùÑ Îã§Î•∏ Î¨∏Ï†úÎ•º Ìë∏Îäî Îç∞ Ïû¨ÏÇ¨Ïö©ÌïòÎäî Í∏∞Î≤ïÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§. ÏòàÏª®ÎåÄ ÎåÄÍ∑úÎ™® ÎßêÎ≠âÏπòÎ•º ÎØ∏Î¶¨ ÌïôÏäµ(pretrain)Ìïú ÏûÑÎ≤†Îî©ÏùÑ Î¨∏ÏÑú Î∂ÑÎ•ò Î™®Îç∏Ïùò ÏûÖÎ†•Í∞íÏúºÎ°ú Ïì∞Í≥†, Ìï¥Îãπ ÏûÑÎ≤†Îî©ÏùÑ Ìè¨Ìï®Ìïú Î™®Îç∏ Ï†ÑÏ≤¥Î•º Î¨∏ÏÑú Î∂ÑÎ•ò Í≥ºÏ†úÎ•º ÏûòÌï† Ïàò ÏûàÎèÑÎ°ù ÏóÖÎç∞Ïù¥Ìä∏(fine-tuning)ÌïòÎäî Î∞©ÏãùÏù¥ Î∞îÎ°ú Í∑∏Í≤ÉÏûÖÎãàÎã§. Î¨ºÎ°† Ï†ÑÏù¥ ÌïôÏäµÏùÄ Î¨∏ÏÑú Î∂ÑÎ•ò Ïù¥Ïô∏Ïùò Îã§ÏñëÌïú Îã§Î•∏ Í≥ºÏ†úÏóêÎèÑ Ï†ÅÏö©Ìï† Ïàò ÏûàÏäµÎãàÎã§.

> Ï∂úÏ≤ò: [ÌïúÍµ≠Ïñ¥ ÏûÑÎ≤†Îî© ÏÑúÎ¨∏](https://ratsgo.github.io/natural%20language%20processing/2019/09/12/embedding/)

#### - Embedding Î†àÏù¥Ïñ¥ 
>- Ïö∞Î¶¨Í∞Ä Í∞ÄÏßÑ ÏÇ¨Ï†ÑÏùò Îã®Ïñ¥ Í∞úÏàò X ÏõåÎìú Î≤°ÌÑ∞ ÏÇ¨Ïù¥Ï¶àÎßåÌÅºÏùò ÌÅ¨Í∏∞Î•º Í∞ÄÏßÑ ÌïôÏäµ ÌååÎùºÎØ∏ÌÑ∞  
>- Í∞êÏÑ± Î∂ÑÎ•ò Î™®Îç∏Ïù¥ ÌïôÏäµÏù¥ Ïûò ÎêòÏóàÎã§Î©¥, Embedding Î†àÏù¥Ïñ¥Ïóê ÌïôÏäµÎêú ÏõåÎìú Î≤°ÌÑ∞Îì§ÎèÑ ÏùòÎØ∏ Í≥µÍ∞ÑÏÉÅÏóê Ïú†ÏùòÎØ∏Ìïú ÌòïÌÉúÎ°ú ÌïôÏäµÎêòÏóàÏùÑ Í≤ÉÏûÑ

```python
embedding_layer = model.layers[0]
weights = embedding_layer.get_weights()[0]
print(weights.shape)    # shape: (vocab_size, embedding_dim)
```

```python
# ÌïôÏäµÌïú Embedding ÌååÎùºÎØ∏ÌÑ∞Î•º ÌååÏùºÏóê Ïç®ÏÑú Ï†ÄÏû•Ìï©ÎãàÎã§. 
word2vec_file_path = os.getenv('HOME')+'/aiffel/sentiment_classification/data/word2vec.txt'
f = open(word2vec_file_path, 'w')
f.write('{} {}\n'.format(vocab_size-4, word_vector_dim))  # Î™áÍ∞úÏùò Î≤°ÌÑ∞Î•º ÏñºÎßà ÏÇ¨Ïù¥Ï¶àÎ°ú Í∏∞Ïû¨Ìï†ÏßÄ ÌÉÄÏù¥ÌãÄÏùÑ ÏîÅÎãàÎã§.

# Îã®Ïñ¥ Í∞úÏàò(ÏóêÏÑú ÌäπÏàòÎ¨∏Ïûê 4Í∞úÎäî Ï†úÏô∏ÌïòÍ≥†)ÎßåÌÅºÏùò ÏõåÎìú Î≤°ÌÑ∞Î•º ÌååÏùºÏóê Í∏∞Î°ùÌï©ÎãàÎã§. 
vectors = model.get_weights()[0]
for i in range(4,vocab_size):
    f.write('{} {}\n'.format(index_to_word[i], ' '.join(map(str, list(vectors[i, :])))))
f.close()
```

#### - Î™®Îç∏Ïùò ÏûÑÎ≤†Îî© Î†àÏù¥Ïñ¥Î•º Word2VecÏùò Í≤ÉÏúºÎ°ú ÍµêÏ≤¥ÌïòÏó¨ Îã§Ïãú ÌïôÏäµÏãúÌÇ§Í∏∞

```python
vocab_size = 10000    # Ïñ¥Ìúò ÏÇ¨Ï†ÑÏùò ÌÅ¨Í∏∞ÏûÖÎãàÎã§(10,000Í∞úÏùò Îã®Ïñ¥)
word_vector_dim = 300  # ÏõåÎìú Î≤°ÌÑ∞Ïùò Ï∞®ÏõêÏàò
embedding_matrix = np.random.rand(vocab_size, word_vector_dim)

# embedding_matrixÏóê Word2Vec ÏõåÎìú Î≤°ÌÑ∞Î•º Îã®Ïñ¥ ÌïòÎÇòÏî©ÎßàÎã§ Ï∞®Î°ÄÏ∞®Î°Ä Ïπ¥ÌîºÌïúÎã§.
for i in range(4,vocab_size):
    if index_to_word[i] in word2vec:
        embedding_matrix[i] = word2vec[index_to_word[i]]
```

```python
from tensorflow.keras.initializers import Constant

vocab_size = 10000    # Ïñ¥Ìúò ÏÇ¨Ï†ÑÏùò ÌÅ¨Í∏∞ÏûÖÎãàÎã§(10,000Í∞úÏùò Îã®Ïñ¥)
word_vector_dim = 300  # ÏõåÎìú Î≤°ÌÑ∞Ïùò Ï∞®Ïõê Ïàò 

# Î™®Îç∏ Íµ¨ÏÑ±
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(vocab_size, 
                                 word_vector_dim, 
                                 embeddings_initializer=Constant(embedding_matrix),  # Ïπ¥ÌîºÌïú ÏûÑÎ≤†Îî©ÏùÑ Ïó¨Í∏∞ÏÑú ÌôúÏö©
                                 input_length=maxlen, 
                                 trainable=True))   # trainableÏùÑ TrueÎ°ú Ï£ºÎ©¥ Fine-tuning
model.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model.add(tf.keras.layers.MaxPooling1D(5))
model.add(tf.keras.layers.Conv1D(16, 7, activation='relu'))
model.add(tf.keras.layers.GlobalMaxPooling1D())
model.add(tf.keras.layers.Dense(8, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid')) 

model.summary()
```

```python
# ÌïôÏäµÏùò ÏßÑÌñâ
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
              
epochs=20  # Î™á epochÎ•º ÌõàÎ†®ÌïòÎ©¥ Ï¢ãÏùÑÏßÄ Í≤∞Í≥ºÎ•º Î≥¥Î©¥ÏÑú Î∞îÍæ∏Ïñ¥ Î¥ÖÏãúÎã§. 

history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)
```

```python
from konlpy.tag import Mecab # ÌïúÍµ≠Ïñ¥ ÌòïÌÉúÏÜå ÌÜ†ÌÅ∞ÌôîÎ•º ÏúÑÌï¥ ÌòïÌÉúÏÜå Î∂ÑÏÑùÍ∏∞ Ï§ë mecab ÏÇ¨Ïö©
import numpy as np
from collections import Counter

tokenizer = Mecab()
stopwords = ['Ïùò','Í∞Ä','Ïù¥','ÏùÄ','Îì§','Îäî','Ï¢Ä','Ïûò','Í±ç','Í≥º','ÎèÑ','Î•º','ÏúºÎ°ú','Ïûê','Ïóê','ÏôÄ','Ìïú','ÌïòÎã§']

def load_data(train_data, test_data, num_words=10000):
    train_data.drop_duplicates(subset=['document'], inplace=True) # Ï§ëÎ≥µ Ï†úÍ±∞
    train_data = train_data.dropna(how='any') # Í≤∞Ï∏°Ïπò ÌïòÎÇòÎùºÎèÑ ÏûàÏúºÎ©¥ Ï†úÍ±∞
    test_data.drop_duplicates(subset=['document'], inplace=True)
    test_data = test_data.dropna(how='any')
    
    X_train = []
    for sentence in train_data['document']:
        temp_X = tokenizer.morphs(sentence) # ÌÜ†ÌÅ∞Ìôî
        temp_X = [word for word in temp_X if not word in stopwords] # Î∂àÏö©Ïñ¥ Ï†úÍ±∞
        X_train.append(temp_X)
        
    X_test = []
    for sentence in test_data['document']:
        temp_X = tokenizer.morphs(sentence) # ÌÜ†ÌÅ∞Ìôî
        temp_X = [word for word in temp_X if not word in stopwords] # Î∂àÏö©Ïñ¥ Ï†úÍ±∞
        X_test.append(temp_X)
    
    words = np.concatenate(X_train).tolist()
    counter = Counter(words) # Í∞Å ÏöîÏÜåÏùò Í∞úÏàò Îã§Î£®Í≥† Ïã∂ÏùÑ Îïå
    counter = counter.most_common(10000-4) # ÎπàÎèÑÏàúÏúºÎ°ú ÎÜíÏùÄ 9996Í∞ú Î¶¨Ïä§Ìä∏ ÏïàÏùò ÌäúÌîåÎ°ú Î∞òÌôò
    vocab = ['<PAD>', '<BOS>', '<UNK>', '<UNUSED>'] + [key for key, _ in counter] # ????
    word_to_index = {word: index for index, word in enumerate(vocab)}
    
    def wordlist_to_indexlist(wordlist): # ÌÖçÏä§Ìä∏ Ïä§Ìä∏ÎßÅÏùÑ ÏÇ¨Ï†Ñ Ïù∏Îç±Ïä§ Ïä§Ìä∏ÎßÅÏúºÎ°ú
        return [word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in wordlist]
    
    X_train = list(map(wordlist_to_indexlist, X_train))
    X_test = list(map(wordlist_to_indexlist, X_test))
    
    return X_train, np.array(list(train_data['label'])), X_test, np.array(list(test_data['label'])), word_to_index
    
    
X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
```

```python
index_to_word = {index:word for word, index in word_to_index.items()}
```

```python
# Î¨∏Ïû• 1Í∞úÎ•º ÌôúÏö©Ìï† ÎîïÏÖîÎÑàÎ¶¨ÏôÄ Ìï®Íªò Ï£ºÎ©¥, Îã®Ïñ¥ Ïù∏Îç±Ïä§ Î¶¨Ïä§Ìä∏ Î≤°ÌÑ∞Î°ú Î≥ÄÌôòÌï¥ Ï£ºÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
# Îã®, Î™®Îì† Î¨∏Ïû•ÏùÄ <BOS>Î°ú ÏãúÏûëÌïòÎäî Í≤ÉÏúºÎ°ú Ìï©ÎãàÎã§. 
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

# Ïó¨Îü¨ Í∞úÏùò Î¨∏Ïû• Î¶¨Ïä§Ìä∏Î•º ÌïúÍ∫ºÎ≤àÏóê Îã®Ïñ¥ Ïù∏Îç±Ïä§ Î¶¨Ïä§Ìä∏ Î≤°ÌÑ∞Î°ú encodeÌï¥ Ï£ºÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# Ïà´Ïûê Î≤°ÌÑ∞Î°ú encodeÎêú Î¨∏Ïû•ÏùÑ ÏõêÎûòÎåÄÎ°ú decodeÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]Î•º ÌÜµÌï¥ <BOS>Î•º Ï†úÏô∏

# Ïó¨Îü¨ Í∞úÏùò Ïà´Ïûê Î≤°ÌÑ∞Î°ú encodeÎêú Î¨∏Ïû•ÏùÑ ÌïúÍ∫ºÎ≤àÏóê ÏõêÎûòÎåÄÎ°ú decodeÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§. 
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]
```
