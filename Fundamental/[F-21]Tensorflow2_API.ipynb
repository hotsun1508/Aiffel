### ğŸ MNIST (1) Sequential API í™œìš©

````python
import tensorflow as tf
from tensorflow import keras
import numpy as np
````

````python
# ë°ì´í„° êµ¬ì„±ë¶€ë¶„
mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train=x_train[...,np.newaxis]  # : = ... ê°™ì€ê±°ì„ !!!
x_test=x_test[...,np.newaxis]

print(len(x_train), len(x_test))
````

````python
# Sequential Modelì„ êµ¬ì„±í•´ì£¼ì„¸ìš”.
"""
Spec:
1. 32ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
2. 64ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
3. Flatten ë ˆì´ì–´
4. 128ê°œì˜ ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ reluì¸ Fully-Connected Layer(Dense)
5. ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì— ë§ëŠ” ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ softmaxì¸ Fully-Connected Layer(Dense)
"""

# ì—¬ê¸°ì— ëª¨ë¸ì„ êµ¬ì„±í•´ì£¼ì„¸ìš”
model = keras.Sequential([
    keras.layers.Conv2D(32, 3, activation='relu'),
    keras.layers.Conv2D(64, 3, activation='relu'),
    keras.layers.Flatten(),
	keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
````

````python
# ëª¨ë¸ í•™ìŠµ ì„¤ì •

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test,  y_test, verbose=2)
````

###ğŸ MNIST (2) Functional API í™œìš©
- `keras.Model`ì„ ì§ì ‘ í™œìš©
- `keras.Input`ìœ¼ë¡œ ì •ì˜ëœ input ë° output ë ˆì´ì–´ êµ¬ì„±ì„ í†µí•´ model êµ¬í˜„

````python
import tensorflow as tf
from tensorflow import keras
import numpy as np
````

````python
mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train=x_train[...,np.newaxis]
x_test=x_test[...,np.newaxis]

print(len(x_train), len(x_test))
````

````python
"""
Spec:
0. (28X28X1) ì°¨ì›ìœ¼ë¡œ ì •ì˜ëœ Input
1. 32ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
2. 64ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
3. Flatten ë ˆì´ì–´
4. 128ê°œì˜ ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ reluì¸ Fully-Connected Layer(Dense)
5. ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì— ë§ëŠ” ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ softmaxì¸ Fully-Connected Layer(Dense)
"""

# ì—¬ê¸°ì— ëª¨ë¸ì„ êµ¬ì„±í•´ ì£¼ì„¸ìš”.
inputs = keras.Input(shape=(28, 28, 1))

x = keras.layers.Conv2D(32, 3, activation='relu')(inputs)
x = keras.layers.Conv2D(64, 3, activation='relu')(x)
x = keras.layers.Flatten()(x)
x = keras.layers.Dense(128, activation='relu')(x)
predictions = keras.layers.Dense(10, activation='softmax')(x)

model = keras.Model(inputs=inputs, outputs=predictions)
````

````python
# ëª¨ë¸ í•™ìŠµ ì„¤ì •

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test,  y_test, verbose=2)
````

##ğŸ MNIST (3) Subclassing í™œìš©
- `Subclassing` ë°©ë²•ì€ `keras.Model`ì„ ìƒì†ë°›ì€ í´ë˜ìŠ¤ë¥¼ ë§Œë“œëŠ” ê²ƒ
- `__init__()` ë©”ì„œë“œ ì•ˆì—ì„œ ë ˆì´ì–´ë¥¼ ì„ ì–¸
- `call()` ë©”ì„œë“œ ì•ˆì—ì„œ `forward propagation`ì„ êµ¬í˜„í•˜ëŠ” ë°©ì‹
- `Functional` ë°©ì‹ê³¼ ë¹„êµí•˜ìë©´, `call()`ì˜ ì…ë ¥ì´ Inputì´ê³ , `call()`ì˜ ë¦¬í„´ê°’ì´ Outputì´ ëœë‹¤

````python
import tensorflow as tf
from tensorflow import keras
import numpy as np
````

````python
# ë°ì´í„° êµ¬ì„±ë¶€ë¶„
mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train=x_train[...,np.newaxis]
x_test=x_test[...,np.newaxis]

print(len(x_train), len(x_test))
````

````python
# Subclassingì„ í™œìš©í•œ Modelì„ êµ¬ì„±í•´ì£¼ì„¸ìš”.
"""
Spec:
0. keras.Model ì„ ìƒì†ë°›ì•˜ìœ¼ë©°, __init__()ì™€ call() ë©”ì„œë“œë¥¼ ê°€ì§„ ëª¨ë¸ í´ë˜ìŠ¤
1. 32ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
2. 64ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
3. Flatten ë ˆì´ì–´
4. 128ê°œì˜ ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ reluì¸ Fully-Connected Layer(Dense)
5. ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì— ë§ëŠ” ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ softmaxì¸ Fully-Connected Layer(Dense)
6. callì˜ ì…ë ¥ê°’ì´ ëª¨ë¸ì˜ Input, callì˜ ë¦¬í„´ê°’ì´ ëª¨ë¸ì˜ Output
"""

# ì—¬ê¸°ì— ëª¨ë¸ì„ êµ¬ì„±í•´ì£¼ì„¸ìš”
class CustomModel(keras.Model):
    def __init__(self):
        super().__init__()
        self.conv1 = keras.layers.Conv2D(32, 3, activation='relu')
        self.conv2 = keras.layers.Conv2D(64, 3, activation='relu')
        self.flatten = keras.layers.Flatten()
        self.fc1 = keras.layers.Dense(128, activation='relu')
        self.fc2 = keras.layers.Dense(10, activation='softmax')
        
    def call(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.fc2(x)
        
        return x
        
model = CustomModel()
````

````python
# ëª¨ë¸ í•™ìŠµ ì„¤ì •

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test,  y_test, verbose=2)
````

````python
# Sequential Modelì„ êµ¬ì„±í•´ì£¼ì„¸ìš”.
"""
Spec:
1. 16ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
2. pool_sizeê°€ 2ì¸ MaxPool ë ˆì´ì–´
3. 32ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
4. pool_sizeê°€ 2ì¸ MaxPool ë ˆì´ì–´
5. 256ê°œì˜ ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ reluì¸ Fully-Connected Layer(Dense)
6. ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì— ë§ëŠ” ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ softmaxì¸ Fully-Connected Layer(Dense)
"""

# ì—¬ê¸°ì— ëª¨ë¸ì„ êµ¬ì„±í•´ì£¼ì„¸ìš”
model = keras.Sequential([
    keras.layers.Conv2D(16, 3, activation='relu'),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Conv2D(32, 3, activation='relu'),
    keras.layers.MaxPool2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(100, activation='softmax')
])
````

````python
# ëª¨ë¸ í•™ìŠµ ì„¤ì •
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# í•™ìŠµ ê´€ë ¨ ë¶€ë¶„ì„ ì‘ì„±í•´ì£¼ì„¸ìš”
model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test,  y_test, verbose=2)
````

##ğŸ CIFAR-100 (2) Functional API í™œìš©
- `keras.Model`ì„ ì§ì ‘ í™œìš©
- `keras.Input`ìœ¼ë¡œ ì •ì˜ëœ input ë° output ë ˆì´ì–´ êµ¬ì„±ì„ í†µí•´ model êµ¬í˜„

````python
import tensorflow as tf
from tensorflow import keras
````

````python
cifar100 = keras.datasets.cifar100

(x_train, y_train), (x_test, y_test) = cifar100.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
print(len(x_train), len(x_test))
````

````python
# Functional APIë¥¼ í™œìš©í•œ Modelì„ êµ¬ì„±í•´ì£¼ì„¸ìš”.
"""
Spec:
0. (32X32X3) ì°¨ì›ìœ¼ë¡œ ì •ì˜ëœ Input
1. 16ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
2. pool_sizeê°€ 2ì¸ MaxPool ë ˆì´ì–´
3. 32ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
4. pool_sizeê°€ 2ì¸ MaxPool ë ˆì´ì–´
5. 256ê°œì˜ ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ reluì¸ Fully-Connected Layer(Dense)
6. ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì— ë§ëŠ” ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ softmaxì¸ Fully-Connected Layer(Dense)
"""

# ì—¬ê¸°ì— ëª¨ë¸ì„ êµ¬ì„±í•´ì£¼ì„¸ìš”
inputs = keras.Input(shape=(32, 32, 3))

x = keras.layers.Conv2D(16, 3, activation='relu')(inputs)
x = keras.layers.MaxPool2D((2,2))(x)
x = keras.layers.Conv2D(32, 3, activation='relu')(x)
x = keras.layers.MaxPool2D((2,2))(x)
x = keras.layers.Flatten()(x)
x = keras.layers.Dense(256, activation='relu')(x)
predictions = keras.layers.Dense(100, activation='softmax')(x)

model = keras.Model(inputs=inputs, outputs=predictions)
````

````python
# ëª¨ë¸ í•™ìŠµ ì„¤ì •
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# í•™ìŠµ ê´€ë ¨ ë¶€ë¶„ì„ ì‘ì„±í•´ì£¼ì„¸ìš”
model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test,  y_test, verbose=2)
````

## ğŸ CIFAR-100 (3) Subclassing í™œìš©
- `Subclassing` ë°©ë²•ì€ `keras.Model`ì„ ìƒì†ë°›ì€ í´ë˜ìŠ¤ë¥¼ ë§Œë“œëŠ” ê²ƒ
- `__init__()` ë©”ì„œë“œ ì•ˆì—ì„œ ë ˆì´ì–´ë¥¼ ì„ ì–¸
- `call()` ë©”ì„œë“œ ì•ˆì—ì„œ `forward propagation`ì„ êµ¬í˜„í•˜ëŠ” ë°©ì‹
- `Functional` ë°©ì‹ê³¼ ë¹„êµí•˜ìë©´, `call()`ì˜ ì…ë ¥ì´ Inputì´ê³ , `call()`ì˜ ë¦¬í„´ê°’ì´ Outputì´ ëœë‹¤

ì—¬ì „íˆ ì •í™•ë„ëŠ” 40% ë¯¸ë§Œì˜ ìˆ˜ì¹˜

````python
import tensorflow as tf
from tensorflow import keras
````

````python
# ë°ì´í„° êµ¬ì„±ë¶€ë¶„
cifar100 = keras.datasets.cifar100

(x_train, y_train), (x_test, y_test) = cifar100.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
print(len(x_train), len(x_test))
````

````python
# Subclassingì„ í™œìš©í•œ Modelì„ êµ¬ì„±í•´ì£¼ì„¸ìš”.
"""
Spec:
0. keras.Model ì„ ìƒì†ë°›ì•˜ìœ¼ë©°, __init__()ì™€ call() ë©”ì„œë“œë¥¼ ê°€ì§„ ëª¨ë¸ í´ë˜ìŠ¤
1. 16ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
2. pool_sizeê°€ 2ì¸ MaxPool ë ˆì´ì–´
3. 32ê°œì˜ ì±„ë„ì„ ê°€ì§€ê³ , ì»¤ë„ì˜ í¬ê¸°ê°€ 3, activation functionì´ reluì¸ Conv2D ë ˆì´ì–´
4. pool_sizeê°€ 2ì¸ MaxPool ë ˆì´ì–´
5. 256ê°œì˜ ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ reluì¸ Fully-Connected Layer(Dense)
6. ë°ì´í„°ì…‹ì˜ í´ë˜ìŠ¤ ê°œìˆ˜ì— ë§ëŠ” ì•„ì›ƒí’‹ ë…¸ë“œë¥¼ ê°€ì§€ê³ , activation functionì´ softmaxì¸ Fully-Connected Layer(Dense)
7. callì˜ ì…ë ¥ê°’ì´ ëª¨ë¸ì˜ Input, callì˜ ë¦¬í„´ê°’ì´ ëª¨ë¸ì˜ Output
"""

# ì—¬ê¸°ì— ëª¨ë¸ì„ êµ¬ì„±í•´ì£¼ì„¸ìš”
class CustomModel(keras.Model):
    def __init__(self):
        super().__init__()
        self.conv1 = keras.layers.Conv2D(16, 3, activation='relu')
        self.maxpool1 = keras.layers.MaxPool2D((2,2))
        self.conv2 = keras.layers.Conv2D(32, 3, activation='relu')
        self.maxpool2 = keras.layers.MaxPool2D((2,2))
        self.flatten = keras.layers.Flatten()
        self.fc1 = keras.layers.Dense(256, activation='relu')
        self.fc2 = keras.layers.Dense(100, activation='softmax')
        
    def call(self, x):
        x = self.conv1(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.maxpool2(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.fc2(x)
        
        return x
        
model = CustomModel()
````

````python
# ëª¨ë¸ í•™ìŠµ ì„¤ì •
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# í•™ìŠµ ê´€ë ¨ ë¶€ë¶„ì„ ì‘ì„±í•´ì£¼ì„¸ìš”
model.fit(x_train, y_train, epochs=5)

model.evaluate(x_test,  y_test, verbose=2)
````

##ğŸ Tensorflowì—ì„œ ì œê³µí•˜ëŠ” `tf.GradientTape`
- ìœ„ì™€ ê°™ì´ ìˆœì „íŒŒ(forward pass) ë¡œ ì§„í–‰ëœ ëª¨ë“  ì—°ì‚°ì˜ ì¤‘ê°„ ë ˆì´ì–´ê°’ì„ **tape**ì— ê¸°ë¡í•˜ê³ , ì´ë¥¼ ì´ìš©í•´ gradientë¥¼ ê³„ì‚°í•œ í›„ **tape**ë¥¼ íê¸°í•˜ëŠ” ê¸°ëŠ¥ì„ ìˆ˜í–‰
- ê·¸ë˜ë””ì–¸íŠ¸ë¥¼ ì¢€ ë” ê³ ê¸‰ìŠ¤ëŸ½ê²Œ í™œìš©í•˜ëŠ” ë‹¤ì–‘í•œ ê¸°ë²•ì„ í†µí•´ ìì£¼ ì‚¬ìš©ë¨

### ì´ì „ ìŠ¤í…ì—ì„œ ì§„í–‰í–ˆë˜ í•™ìŠµì„ `tf.GradientTape`ë¥¼ ì´ìš©í•œ ê²ƒìœ¼ë¡œ ë³€í˜•í•´ ë³´ì. 

````python
import tensorflow as tf
from tensorflow import keras

# ë°ì´í„° êµ¬ì„±ë¶€ë¶„
cifar100 = keras.datasets.cifar100

(x_train, y_train), (x_test, y_test) = cifar100.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
print(len(x_train), len(x_test))

# ëª¨ë¸ êµ¬ì„±ë¶€ë¶„
class CustomModel(keras.Model):
    def __init__(self):
        super().__init__()
        self.conv1 = keras.layers.Conv2D(16, 3, activation='relu')
        self.maxpool1 = keras.layers.MaxPool2D((2,2))
        self.conv2 = keras.layers.Conv2D(32, 3, activation='relu')
        self.maxpool2 = keras.layers.MaxPool2D((2,2))
        self.flatten = keras.layers.Flatten()
        self.fc1 = keras.layers.Dense(256, activation='relu')
        self.fc2 = keras.layers.Dense(100, activation='softmax')

    def call(self, x):
        x = self.conv1(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.maxpool2(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.fc2(x)

        return x

model = CustomModel()
````

ì—¬ê¸°ê¹Œì§€ëŠ” ì•ì—ì„œ ë‹¤ë£¨ì—ˆë˜ Subclassingì„ í™œìš©í•œ ëª¨ë¸ ì‘ì„±ë²•ê³¼ ì „í˜€ ë‹¤ë¥´ì§€ ì•ŠìŠµë‹ˆë‹¤. ë‹¬ë¼ì§€ëŠ” ê²ƒì€ model.compile(), model.fit()ì„ í†µí•´ ì†ì‰½ê²Œ ì§„í–‰í–ˆë˜ í•™ìŠµ ì„¸íŒ… ë° ìˆ˜í–‰ ë¶€ë¶„ì…ë‹ˆë‹¤.

- **model.compile()**: ëª¨ë¸ í•™ìŠµì„ ìœ„í•´ `loss`, `optimizer`ë¥¼ ì§€ì •í•´ ì£¼ë©´ ë‚´ë¶€ì ìœ¼ë¡œëŠ” ë§¤ ìŠ¤í… í•™ìŠµì´ ì§„í–‰ë  ë•Œë§ˆë‹¤ ë°œìƒí•˜ëŠ” loss ë° ê·¸ë˜ë””ì–¸íŠ¸ê°€ ì–´ë–»ê²Œ í•™ìŠµ íŒŒë¼ë¯¸í„°ë¥¼ ì—…ë°ì´íŠ¸í•˜ê²Œ ë˜ëŠ”ì§€ë¥¼ ì§€ì •í•´ ì£¼ëŠ” ì‘ì—…ì´ `model.compile()` ì•ˆì—ì„œ ìë™ìœ¼ë¡œ ì§„í–‰

### ğŸ ì•„ë˜ ì½”ë“œ
- `tape.gradient()`ë¥¼ í†µí•´ ë§¤ ìŠ¤í… í•™ìŠµì´ ì§„í–‰ë  ë•Œë§ˆë‹¤ ë°œìƒí•˜ëŠ” ê·¸ë˜ë””ì–¸íŠ¸ë¥¼ ì¶”ì¶œí•œ í›„ 
- `optimizer.apply_gradients()`ë¥¼ í†µí•´ ë°œìƒí•œ ê·¸ë˜ë””ì–¸íŠ¸ê°€ 
- ì—…ë°ì´íŠ¸í•´ì•¼ í•  íŒŒë¼ë¯¸í„° `model.trainable_variables`ë¥¼ ì§€ì •í•´ ì£¼ëŠ” ê³¼ì •ì„ ê¸°ìˆ í•œ ê²ƒì…ë‹ˆë‹¤.

````python
loss_func = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.Adam()

# tf.GradientTape()ë¥¼ í™œìš©í•œ train_step
def train_step(features, labels):
    with tf.GradientTape() as tape:  # withë¡œ GradientTape()ë¥¼ tapeë¡œ ì ê¹ ì‚¬ìš© í›„ ë°˜ë‚© 
        predictions = model(features)
        loss = loss_func(labels, predictions)
        gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))  # zip ì••ì¶•
    return loss
````

> ë§¤ ìŠ¤í… ì§„í–‰ë˜ëŠ” í•™ìŠµì˜ ì‹¤ì œ ë™ì‘ì´ `train_step()` ë©”ì„œë“œë¡œ êµ¬í˜„ë˜ì—ˆìŠµë‹ˆë‹¤.
```
model.fit(x_train, y_train, epochs=5, batch_size=32)
```
### `model.fit()`ìœ¼ë¡œ ìœ„ì™€ ê°™ì´ í•œ ì¤„ë¡œ ê°„ë‹¨í•˜ê²Œ ìˆ˜í–‰ë˜ë˜ ì‹¤ì œ ë°°ì¹˜ í•™ìŠµ ê³¼ì •
- ë§¤ ìŠ¤í…ë§ˆë‹¤ ìœ„ì—ì„œ êµ¬í˜„í–ˆë˜ `train_step()`ê°€ í˜¸ì¶œë˜ëŠ” ê³¼ì •ìœ¼ë¡œ ë°”ê¾¸ì–´ êµ¬í˜„ ê°€ëŠ¥. 
- `model.fit()` í˜¸ì¶œ ì‹œì— ê²°ì •ë˜ëŠ” `batch_size`ë§Œ ê²°ì •í•´ ì£¼ë©´ ëœë‹¤.



````python
import time
def train_model(batch_size=32):
    start = time.time()
    for epoch in range(5):
        x_batch = []
        y_batch = []
        for step, (x, y) in enumerate(zip(x_train, y_train)):  # x_train, y_train ìŒì˜ í¬ê¸°ë§Œí¼ì˜ stepë§ˆë‹¤ (x_trainì´ë‘ y_train)ì´ ìŒìœ¼ë¡œ append
            x_batch.append(x)
            y_batch.append(y)
            if step % batch_size == batch_size-1:  # ì´ê±¸ ë§Œì¡±í•˜ëŠ” ê²½ìš° x_batch, y_batch ì´ˆê¸°í™” // step =0ë¶€í„° ì‹œì‘ batch size 0~31ê¹Œì§€ ê°€ì•¼ì§€ë§Œ batch_size=32ê°€ ë˜ë‹ˆê¹Œ..ã…‡ã…‡
                loss = train_step(np.array(x_batch, dtype=np.float32), np.array(y_batch, dtype=np.float32)) # epochë³„ë¡œ lossê°’ì„ ê³„ì‚°í•˜ê¸° ìœ„í•´ í•„ìš”í•œ ì½”ë“œ
                x_batch = []
                y_batch = []
        print('Epoch %d: last batch loss = %.4f' % (epoch, float(loss))) 
    print("It took {} seconds".format(time.time() - start))

train_model()
````

> #### ğŸ ìœ„ì—ì„œ êµ¬í˜„í•œ `train_model()` ë©”ì„œë“œê°€ ì‹¤ì€ ìš°ë¦¬ê°€ ê·¸ë™ì•ˆ ì‚¬ìš©í–ˆë˜ `model.fit()` ë©”ì„œë“œì™€ ê¸°ëŠ¥ì ìœ¼ë¡œ ê°™ë‹¤.

- ì´ë ‡ë“¯ `tf.GradientTape()`ë¥¼ í™œìš©í•˜ë©´ `model.compile()`ê³¼ `model.fit()` ì•ˆì— ê°ì¶”ì–´ì ¸ ìˆë˜ í•œ ìŠ¤í…ì˜ í•™ìŠµ ë‹¨ê³„(ìœ„ ì˜ˆì œì—ì„œëŠ” `train_step` ë©”ì„œë“œ)ë¥¼ ë„ì§‘ì–´ë‚´ì„œ ììœ ë¡­ê²Œ ì¬êµ¬ì„±
- ì§€ë„í•™ìŠµ ë°©ì‹ê³¼ ë‹¤ë¥¸ ê°•í™”í•™ìŠµ ë˜ëŠ” GAN(Generative Advasarial Network)ì˜ í•™ìŠµì„ ìœ„í•´ì„œëŠ” `train_step` ë©”ì„œë“œì˜ ì¬êµ¬ì„±ì´ í•„ìˆ˜ì ì´ë¯€ë¡œ `tf.GradientTape()`ì˜ í™œìš©ë²• ìˆ™ì§€ í•„ìš” !!
